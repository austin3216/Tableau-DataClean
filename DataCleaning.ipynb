```python
# Import pandas

import pandas as pd
```

```python
# Read Feb csv for data format to match Jan to and pull station IDs

feb_df = pd.read_csv("/Users/amber/Desktop/GitHub/Tableau-DataClean/csvs/JC-202102-citibike-tripdata.csv")
feb_df
```

```python
# Pull start station IDs and Names

feb_df_station = feb_df.loc[:, 'start_station_name':'start_station_id']
feb_df_station = feb_df_station.drop_duplicates()
feb_df_station.sort_values(by='start_station_id')
feb_df_station.reset_index(drop=True, inplace=True)
feb_df_station.to_csv(r'/Users/amber/Desktop/GitHub/Tableau-DataClean/csvs/start_station_ids.csv', index=False)
```

```python
# Pull end station IDs and Names

feb_df_station_end = feb_df.loc[:, 'end_station_name':'end_station_id']
feb_df_station_end = feb_df_station_end.drop_duplicates()
feb_df_station_end.reset_index(drop=True, inplace=True)
feb_df_station_end.to_csv(r'/Users/amber/Desktop/GitHub/Tableau-DataClean/csvs/end_station_ids.csv', index=False)
```

```python
feb_df_station.to_dict('list')
```

```python
# Read in Jan data to clean

import pandas as pd

jan_df = pd.read_csv("/Users/amber/Desktop/GitHub/Tableau-DataClean/csvs/JC-202101-citibike-tripdata.csv")
jan_df
```

```python
# Clean column names

jan_df.columns = jan_df.columns.str.replace(' ', '_')
jan_df
```

```python
# Bring in start station IDs

new_jan_df = pd.merge(jan_df, feb_df_station, on='start_station_name')
new_jan_df
```

```python
# Bring in end station IDs

new_jan_df = pd.merge(new_jan_df, feb_df_station_end, on='end_station_name')
new_jan_df
```

```python
# Drop unnecessary columns

new_jan_df.drop(columns=['start_station_id_x', 'end_station_id_x'], inplace=True)
new_jan_df
```

```python
# Rename columns

new_jan_df.rename(columns = {'start_station_id_y':'start_station_id', 'end_station_id_y':'end_station_id'}, inplace=True)
new_jan_df
```

```python
# Reorder columns, push to csv

new_jan_df = new_jan_df.reindex(columns=['tripduration', 'starttime', 'stoptime', 'start_station_id', 'start_station_name', 'start_station_latitude', 'start_station_longitude', 'end_station_id', 'end_station_name', 'end_station_latitude', 'end_station_longitude', 'bikeid', 'usertype', 'birth_year', 'gender'])
new_jan_df.to_csv(r'/Users/amber/Desktop/GitHub/Tableau-DataClean/csvs/JC-202101-citibike-tripdata-CLEANED.csv', index=False)
new_jan_df
```

```python
# Updating usertype to match Feb-Mar data -- making assumption that Subscriber = Member and Customer = Casual
new_jan_df['usertype'] = new_jan_df['usertype'].replace(['Subscriber','Customer'], ['member', 'casual'])
new_jan_df
```

```python
# Renamed additional columns to match Feb-Mar data

new_jan_df.rename(columns={'starttime':'started_at', 'stoptime':'ended_at', 'start_station_latitude':'start_lat', 'start_station_longitude':'start_lng', 'end_station_latitude':'end_lat', 'end_station_longitude':'end_lng', 'usertype':'member_casual'}, inplace=True)
new_jan_df.to_csv(r'/Users/amber/Desktop/GitHub/Tableau-DataClean/csvs/JC-202101-citibike-tripdata-CLEANED.csv', index=False)
new_jan_df
```

```python
# Read in Mar data

mar_df = pd.read_csv("/Users/amber/Desktop/GitHub/Tableau-DataClean/csvs/JC-202103-citibike-tripdata.csv")
mar_df
```

```python
# Combine Feb and Mar data

feb_mar_df = pd.concat([feb_df, mar_df]).reset_index(drop=True)
feb_mar_df.to_csv(r'/Users/amber/Desktop/GitHub/Tableau-DataClean/csvs/FEB-MAR-citibike-tripdata.csv', index=False)
feb_mar_df
```

```python
# Combine Jan and Feb-Mar data for 3-Month span

jan_mar_df = pd.concat([new_jan_df, feb_mar_df]).reset_index(drop=True)
jan_mar_df
```

```python
# Drop columns that won't be used - not used across months

jan_mar_df.drop(['tripduration', 'bikeid', 'birth_year', 'gender', 'ride_id', 'rideable_type'], axis=1, inplace=True)
jan_mar_df
```

```python
# Clean up Station Names to use

jan_mar_df = jan_mar_df.replace('\&','and',regex=True)
jan_mar_df
```

```python
jan_mar_df['start_station_name'] = jan_mar_df['start_station_name'].str.replace(" ","-")
jan_mar_df['end_station_name'] = jan_mar_df['end_station_name'].str.replace(" ","-")
jan_mar_df
```

```python
# create path_id field for future spider map
jan_mar_df['path_id'] = jan_mar_df['start_station_name'] + '_' + jan_mar_df['end_station_name']
jan_mar_df['path_id_2'] = jan_mar_df['start_station_id'] + '_' + jan_mar_df['end_station_id']
jan_mar_df
```

```python
# standardize lat and lon coordinates - was getting layered values in Tableau visualization due to slight variations in lat/lng
jan_mar_df = jan_mar_df.round(3)
jan_mar_df.to_csv(r'/Users/amber/Desktop/GitHub/Tableau-DataClean/csvs/JAN-MAR-citibike-tripdata.csv', index=False)
```

```python
# split dataframes to create origin-destination rows to create dataset for spider map
origin_df = jan_mar_df[[
'started_at', 'ended_at', 'start_station_name', 'start_station_id', 'start_lat', 'start_lng', 'member_casual', 'path_id', 'path_id_2']].copy()
origin_df
```

```python
destination_df = jan_mar_df[[
'started_at', 'ended_at', 'end_station_name', 'end_station_id', 'end_lat', 'end_lng', 'member_casual', 'path_id', 'path_id_2']].copy()
destination_df
```

```python
origin_df['origin_destination'] = 'origin'
destination_df['origin_destination'] = 'destination'
```

```python
# origin_df
destination_df
```

```python
# Make column names consistent

origin_df.rename(columns={"start_station_name":"station_name", "start_station_id":"station_id", "start_lat":"latitude", "start_lng":"longitude"}, inplace=True)
origin_df
```

```python
destination_df.rename(columns={"end_station_name":"station_name", "end_station_id":"station_id", "end_lat":"latitude", "end_lng":"longitude"}, inplace=True)
destination_df
```

```python
# concatenate dataframes

frames = [origin_df, destination_df]

spider_map_df = pd.concat(frames, sort=False)
spider_map_df
```

```python
# Sort values by shared attributes to create (2) lat/lng records per trip - (1) origin, (1) destination

spider_map_df.sort_values(["started_at", "ended_at"], ascending=[True, True], inplace=True)
spider_map_df
```

```python
spider_map_df.to_csv(r'/Users/amber/Desktop/GitHub/Tableau-DataClean/csvs/JAN-MAR-SPIDER-citibike-tripdata.csv', index=False)
```

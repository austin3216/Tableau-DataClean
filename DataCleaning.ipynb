```python
import pandas as pd
```

```python
feb_df = pd.read_csv("/Users/amber/Desktop/GitHub/Tableau-DataClean/csvs/JC-202102-citibike-tripdata.csv")
feb_df
```

```python
feb_df_station = feb_df.loc[:, 'start_station_name':'start_station_id']
feb_df_station = feb_df_station.drop_duplicates()
feb_df_station.sort_values(by='start_station_id')
feb_df_station.reset_index(drop=True, inplace=True)
feb_df_station.to_csv(r'/Users/amber/Desktop/GitHub/Tableau-DataClean/csvs/start_station_ids.csv', index=False)
```

```python
feb_df_station_end = feb_df.loc[:, 'end_station_name':'end_station_id']
feb_df_station_end = feb_df_station_end.drop_duplicates()
feb_df_station_end.reset_index(drop=True, inplace=True)
feb_df_station_end.to_csv(r'/Users/amber/Desktop/GitHub/Tableau-DataClean/csvs/end_station_ids.csv', index=False)
```

```python
import pandas as pd

jan_df = pd.read_csv("/Users/amber/Desktop/GitHub/Tableau-DataClean/csvs/JC-202101-citibike-tripdata.csv")
jan_df
```

```python
jan_df.columns = jan_df.columns.str.replace(' ', '_')
jan_df
```

```python
new_jan_df = pd.merge(jan_df, feb_df_station, on='start_station_name')
new_jan_df
```

```python
new_jan_df = pd.merge(new_jan_df, feb_df_station_end, on='end_station_name')
new_jan_df
```

```python
new_jan_df.drop(columns=['start_station_id_x', 'end_station_id_x'], inplace=True)
new_jan_df
```

```python
new_jan_df.rename(columns = {'start_station_id_y':'start_station_id', 'end_station_id_y':'end_station_id'}, inplace=True)
new_jan_df
```

```python
new_jan_df = new_jan_df.reindex(columns=['tripduration', 'starttime', 'stoptime', 'start_station_id', 'start_station_name', 'start_station_latitude', 'start_station_longitude', 'end_station_id', 'end_station_name', 'end_station_latitude', 'end_station_longitude', 'bikeid', 'usertype', 'birth_year', 'gender'])
new_jan_df.to_csv(r'/Users/amber/Desktop/GitHub/Tableau-DataClean/csvs/JC-202101-citibike-tripdata-CLEANED.csv', index=False)
new_jan_df
```

```python
# Updating usertype to match Feb-Jun data -- making assumption that Subscriber = Member and Customer = Casual
new_jan_df['usertype'] = new_jan_df['usertype'].replace(['Subscriber','Customer'], ['member', 'casual'])
new_jan_df
```
